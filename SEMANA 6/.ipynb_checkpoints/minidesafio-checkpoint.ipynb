##### Obtener información de las tendencias asociadas a los siguientes términos: "Fin del mundo", "Calentamiento global", "Terremoto",'Tsunami'
##### Analizar qué representa el valor obtenido por Pytrends
##### 
Generar un gráfico comparativo donde puedas visualizar como ha sido la evolución de estos términos a lo largo del pasado reciente (últimos 5 años 



```python
!pip install pytrends
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq()
keywords = ["Fin del mundo", "Calentamiento global", "Terremoto",'Tsunami']
pytrends.build_payload(keywords, cat=0, geo='', gprop='') # Datos de los ultimos 5 años
stop_queries = pytrends.interest_over_time()[keywords]
stop_queries.head()
```

```python
stop_queries.plot(kind='line',figsize=(12,6), xlabel='Fecha',ylabel='Interés de audiencia',title='Interés en el tiempo')
```
