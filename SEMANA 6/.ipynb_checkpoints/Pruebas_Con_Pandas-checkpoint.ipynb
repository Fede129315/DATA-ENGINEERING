## Ejercicio Clase

```python
# importo pandas
import pandas as pd
```

```python
# leo el archivo CSV y añado na_values = '$null$' para indicarle como quiero que se vean los null
pdf = pd.read_csv("cars_clus.csv", na_values = '$null$')
pdf
```

```python
# con la funcion dropna() elimino los registros con nulls
pdf = pdf.dropna()
pdf
```

```python
#hacemos un group by de marca y precio (promedio) con mean
pdf_manufact = pdf[["manufact","price"]].groupby(by = "manufact").mean()
pdf_manufact
```

```python
# Agrego una columna con assign 
pdf_manufact = pdf_manufact.assign(valor_nuevo = lambda x: x["price"]*1.05)
pdf_manufact
```

```python
# con describe trae varias estadisticas sobre las columnas
pdf[["manufact","price"]].groupby(by="manufact").describe()
```
