```python
import pandas as pd

datos_reunion=pd.read_csv('microdesafio.csv',sep=';')
datos_reunion
```

```python
from anonymizedf.anonymizedf import anonymize
# Prepare the data to be anonymized
anon = anonymize(datos_reunion)
# Select what data you want to anonymize and your preferred style
anon.fake_names("Comisionado", chaining=True)
anon.fake_dates("Fecha", chaining=True)
anon.fake_whole_numbers("Telefono", chaining=True)

print(datos_reunion.columns)


datos_a_guardar = datos_reunion[['Pais ','Fake_Comisionado','Reduccion_CO2', 'Incrmento_P',
       'Inversion_arboles','Fake_Fecha', 'Fake_Telefono']]
datos_a_guardar.columns = ['Pais ','Comisionado','Reduccion_CO2', 'Incrmento_P',
       'Inversion_arboles','Fecha', 'Telefono']
datos_a_guardar
```

```python
import numpy as np

for i in range(len(datos_a_guardar.index)):
    list_item = []
    for j in range(7):
        if type(list(datos_a_guardar.iloc[i])[j]) is np.int64:
            list_item.append(list(datos_a_guardar.iloc[i])[j].item())
        else:
            list_item.append(list(datos_a_guardar.iloc[i])[j])
    print(list_item)
```
